#### Ejemplos de if, if else, if elif

Ejemplo if

```python
edad = 20                      # Asigna el valor 20 a la variable edad [8]
if edad >= 18:                 # Verifica si edad es mayor o igual a 18 (operador relacional) [9]
    print("Puedes votar.")     # Imprime el mensaje si la condición es verdadera [10]
```

```python
edad = 16                      # Asigna el valor 16 a la variable edad
if edad >= 18:                 # Verifica si edad es mayor o igual a 18 (operador relacional >=)
    print("Puedes votar.")     # Se ejecuta si la condición es verdadera
else:                          # Si la condición anterior es falsa
    print("No puedes votar.")  # Se ejecuta si la edad es menor a 18
```

```python
numero = 6  # Define una variable con un valor entero [11]
if numero % 2 == 0:  # Evalúa si el resto de dividir por 2 es cero (operadores % y ==) [4]
    print("El número es par.")  # Se ejecuta si la división es exacta [13]
else:  # Si el resto no fue cero [1]
    print("El número es impar.")  # Mensaje para los casos que no cumplen el 'if' [13]
```

```python
nota =100  # Define la calificación obtenida [8]
if nota >= 90 and nota <= 100:  # Verifica rango 90-100 usando el operador lógico 'and' [15]
    print("Excelente")  # Mensaje para notas altas
elif nota >= 80 and nota < 90:  # Evalúa el siguiente rango si el anterior falló [1]
    print("Muy bueno")  # Mensaje para notas entre 80 y 89
elif nota >= 0 and nota < 80:  # Tercera condición para notas menores [16]
    print("Necesita mejorar")  # Mensaje de retroalimentación
else:  # Si no cae en ninguno de los rangos anteriores (ej. nota negativa o mayor a 100) [1]
    print("Calificación no válida")  # Mensaje de error
```
